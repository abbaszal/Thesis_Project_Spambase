### Nash Equilibrium Occurrences with Low-Quality Clients

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_nash_equilibrium(csv_filename, algorithm_name, dataset_name):
    df = pd.read_csv(csv_filename, dtype={"Nash Equilibrium": str})  
    df["Total Occurrences"] = df.iloc[:, 1:].sum(axis=1)
    df = df.sort_values(by="Total Occurrences", ascending=False)
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    x_labels = df["Nash Equilibrium"].astype(str)
    x = np.arange(len(x_labels))
    bar_width = 0.15
    
    colors = ["blue", "green", "red", "purple", "orange"]
    categories = df.columns[1:-1]
    
    for i, category in enumerate(categories):
        ax.bar(x + i * bar_width, df[category], width=bar_width, label=category, color=colors[i % len(colors)])
    
    ax.set_xlabel("Nash Equilibrium")
    ax.set_ylabel("Total Occurrences")
    ax.set_title(f"Nash Equilibrium Occurrences with {algorithm_name} on {dataset_name} With Low-Quality Clients")
    
    ax.set_xticks(x + bar_width * 2)
    ax.set_xticklabels(x_labels, rotation=30, ha="right", fontsize=10)
    
    ax.legend()
    plt.show()
```

```python
csv_filename = "/.../plots/FedLR_HuGaDB_With_LQC.csv"
algorithm_name ="FedLR"
dataset_name = "HuGaDB"
plot_nash_equilibrium(csv_filename, algorithm_name, dataset_name)
```

```python
csv_filename = "/.../plots/FedFor_HuGaDB_With_LQC.csv"
algorithm_name ="FedFor"
dataset_name = "HuGaDB"
plot_nash_equilibrium(csv_filename, algorithm_name, dataset_name)
```

```python
csv_filename = "/.../plots/FedLR_Spambase_With_LQC.csv"
algorithm_name ="FedLR"
dataset_name = "Spambase"
plot_nash_equilibrium(csv_filename, algorithm_name, dataset_name)
```

```python
csv_filename = "/.../plots/FedFor_Spambase_With_LQC.csv"
algorithm_name ="FedFor"
dataset_name = "Spambase"
plot_nash_equilibrium(csv_filename, algorithm_name, dataset_name)
```

### Nash Equilibrium Occurrences without Low-Quality Clients

```python
csv_filename = "/.../plots/FedLR_Without_LQC.csv"
df = pd.read_csv(csv_filename, dtype={"Nash Equilibrium": str})


fig, ax = plt.subplots(figsize=(12, 6))


x_labels = df["Nash Equilibrium"]
x = np.arange(len(x_labels))
bar_width = 0.4


ax.bar(x - bar_width/2, df["FedLR_HuGaDB"], width=bar_width, label="FedLR_HuGaDB", color="skyblue", edgecolor="blue", alpha=0.7)
ax.bar(x + bar_width/2, df["FedLR_Spambase"], width=bar_width, label="FedLR_Spambase", color="salmon", edgecolor="red", alpha=0.7)


total_trials_hugadb = 80
total_trials_spambase = 20

ax.axhline(y=total_trials_hugadb, linestyle="dashed", color="blue", label=f"Total Trials FedLR_HuGaDB: {total_trials_hugadb}")
ax.axhline(y=total_trials_spambase, linestyle="dashed", color="red", label=f"Total Trials FedLR_Spambase: {total_trials_spambase}")


ax.set_xlabel("Nash Equilibrium")
ax.set_ylabel("Total Occurrences")
ax.set_title("Nash Equilibrium Occurrences with FedLR on HuGaDB and Spambase without LQC")


ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=30, ha="right", fontsize=10)


ax.legend(loc="upper right", bbox_to_anchor=(1,1))


plt.tight_layout()


plt.show()

```
